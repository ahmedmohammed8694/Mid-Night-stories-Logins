```python
import sqlite3
import os

db_path = r'd:\My Applications\Midnigth stories\data\stories.db'
print("DB exists:", os.path.exists(db_path))

conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Get table schema for books
cursor.execute("PRAGMA table_info(books)")
print("Books columns:")
for row in cursor.fetchall():
    print(row)

# Get table schema for categories
cursor.execute("PRAGMA table_info(categories)")
print("\nCategories columns:")
for row in cursor.fetchall():
    print(row)

# Get books records
cursor.execute("SELECT id, title, file_url, status FROM books")
print("\nBooks list:")
for row in cursor.fetchall():
    print(row)

conn.close()
```

```python
import sqlite3
db_path = r'd:\My Applications\Midnigth stories\data\stories.db'
conn = sqlite3.connect(db_path)
cursor = conn.cursor()
try:
    cursor.execute("""
        SELECT b.*,
          (SELECT GROUP_CONCAT(c.name) FROM book_categories bc JOIN categories c ON bc.category_id = c.id WHERE bc.book_id = b.id) as category_names,
          (SELECT GROUP_CONCAT(bc.category_id) FROM book_categories bc WHERE bc.book_id = b.id) as category_ids,
          (SELECT GROUP_CONCAT(t.name) FROM book_tags bt JOIN tags t ON bt.tag_id = t.id WHERE bt.book_id = b.id) as tag_names
        FROM books b
        WHERE b.id = ?
    """, (1,))
    print("Result:", cursor.fetchone())
except Exception as e:
    print("SQL Error:", e)
conn.close()
```

```python
import urllib.request
try:
    url = "http://localhost:3000/api/books?channel=education"
    req = urllib.request.Request(url)
    with urllib.request.urlopen(req) as response:
        print("STATUS:", response.status)
        print("BODY:", response.read().decode('utf-8'))
except urllib.error.HTTPError as e:
    print("HTTP Error Status:", e.code)
    print("HTTP Error Body:", e.read().decode('utf-8'))
except Exception as e:
    print("Error:", e)
```
