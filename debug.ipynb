```python
import sqlite3
import os

db_path = r'd:\My Applications\Midnigth stories\data\stories.db'
print("DB exists:", os.path.exists(db_path))

conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Get table schema for books
cursor.execute("PRAGMA table_info(books)")
print("Books columns:")
for row in cursor.fetchall():
    print(row)

# Get table schema for categories
cursor.execute("PRAGMA table_info(categories)")
print("\nCategories columns:")
for row in cursor.fetchall():
    print(row)

# Get books records
cursor.execute("SELECT id, title, file_url, status FROM books")
print("\nBooks list:")
for row in cursor.fetchall():
    print(row)

conn.close()
```
